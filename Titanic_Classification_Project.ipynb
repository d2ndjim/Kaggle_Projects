### The ML project is built using Kaggle's Titanic Dataset. The ML model used is Logistic Regression which predicts the survival of passengers on the Titanic Ship.

```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#import training datasets
passengers = pd.read_csv('train.csv')
passengers.head()
```

```python
#inspect data 
passengers.info()
```

### The next step is to select columns that are relevant to our model prediction. This is based on common logic. Examples of this columns will be PCLASS as a passengers class might determine if the passenger gets to leave the boat on time or not, SEX as women might be considered to leave the boat before men, AGE as children might be considered to leave the boat before adults and SURVIVED as this determines if a passenger survived or not

```python
columns_to_remove = ['PassengerId','Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked']
passengers.drop(columns_to_remove, axis=1, inplace=True)
passengers.head()
```

```python
#clean_data steps
# 1. fill missing age values with the mean of ages
# 2. fill sex with numerical categorical variables i.e 1 for male and 0 for females
# 3. group the passengers classes seperately
passengers['Age'].fillna(value = round(passengers['Age'].mean()),inplace=True)
passengers['Sex'] = passengers['Sex'].map({'male': 1, 'female': 0})
passengers['First_Class'] = passengers['Pclass'].apply(lambda x: 1 if x == 1 else 0)
passengers['Second_Class'] = passengers['Pclass'].apply(lambda x: 1 if x == 2 else 0)
passengers.head()
```

```python
passengers.drop(['Pclass'], axis=1, inplace=True)
```

```python
passengers.head()
```

```python
#select the independent variables(features) and the dependent variable(label)
features = passengers[['Sex','Age','First_Class','Second_Class']]
label = passengers['Survived']
#syntax for splitting data into train and test sets - X_train,X_test,Y_train,Y_test = train_test_split(features,label)
features_train,features_test,label_train,label_test = train_test_split(features,label,train_size = .8)

```

```python
scaler = StandardScaler()
#scale data
scaler.fit_transform(features_train)
scaler.transform(features_test)
```

```python
#model building
#name logistic regression as an object
model = LogisticRegression()
model.fit(features,label)
sorted(list(zip(['Sex','Age','First_Class','Second_Class'],model.coef_)),key = lambda x: abs(x[1]),reverse=True)
```

```python
accuracy = model.score(features_train,label_train)
print(accuracy*100)
```

```python
Lekan = np.array([1.0,10.0,1.0,0.0])
Kenny = np.array([0.0,250.0,0.0,0.0])

lekan_kenny = np.array([Lekan,Kenny])
sample_prediction = scaler.transform(lekan_kenny)
```

```python
pred = model.predict(sample_prediction)
print(pred)
```

```python
columns_to_remove2 = ['Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked']
passengers2.drop(columns_to_remove2, axis=1, inplace=True)
passengers2.head()
```

```python
pred_data = scaler.transform(pred_features)
predictions = model.predict(pred_data)
print(predictions)
```

```python
TitanicPrediction_file = pd.DataFrame({"PassengerId": passengers2["PassengerId"], "Survived": predictions})
TitanicPrediction_file.to_csv('TitanicPrediction_file.csv', index=False)
TitanicPrediction_file
```
