```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#import training datasets
passengers = pd.read_csv('train.csv')
passengers.head()
```

```python
#inspect data 
passengers.info()
```

```python
columns_to_remove = ['PassengerId','Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked']
passengers.drop(columns_to_remove, axis=1, inplace=True)
passengers.head()
```

```python
#clean_data steps
# 1. fill missing age values with the mean of ages
# 2. fill sex with numerical categorical variables i.e 1 for male and 0 for females
# 3. group the passengers classes seperately
passengers['Age'].fillna(value = round(passengers['Age'].mean()),inplace=True)
passengers['Sex'] = passengers['Sex'].map({'male': 1, 'female': 0})
passengers['First_Class'] = passengers['Pclass'].apply(lambda x: 1 if x == 1 else 0)
passengers['Second_Class'] = passengers['Pclass'].apply(lambda x: 1 if x == 2 else 0)
passengers.head()
```

```python
passengers.drop(['Pclass'], axis=1, inplace=True)
passengers.head()
```

```python
#select the independent variables(features) and the dependent variable(label)
features = passengers[['Sex','Age','First_Class','Second_Class']]
label = passengers['Survived']
#syntax for splitting data into train and test sets - X_train,X_test,Y_train,Y_test = train_test_split(features,label)
training_data,validation_data,training_labels,validation_labels = train_test_split(features,label,train_size = 0.8)
```

```python
classifier = SVC(gamma=1, C=2)
classifier.fit(training_data,training_labels)
score = classifier.score(validation_data,validation_labels)
```

```python
Lekan = np.array([1.0,10.0,1.0,0.0])
Kenny = np.array([0.0,250.0,0.0,0.0])

lekan_kenny = np.array([Lekan,Kenny])
pred = classifier.predict([Lekan,Kenny])
print(pred)
```

```python
#import prediction dataset
passengers2 = pd.read_csv('test.csv')
passengers2
```

```python
columns_to_remove2 = ['Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked']
passengers2.drop(columns_to_remove2, axis=1, inplace=True)
passengers2.head()
```

```python
#clean prediction data
passengers2['Age'].fillna(value = round(passengers2['Age'].mean()),inplace=True)
passengers2['Sex'] = passengers2['Sex'].map({'male': 1, 'female': 0})
passengers2['First_Class'] = passengers2['Pclass'].apply(lambda x: 1 if x == 1 else 0)
passengers2['Second_Class'] = passengers2['Pclass'].apply(lambda x: 1 if x == 2 else 0)
passengers2.drop(['Pclass'], axis=1, inplace=True)
pred_features = passengers2[['Sex','Age','First_Class','Second_Class']]
```

```python
predictions = classifier.predict(pred_features)
print(predictions)
```

```python
TitanicPrediction = pd.DataFrame({"PassengerId": passengers2["PassengerId"], "Survived": predictions})
TitanicPrediction.to_csv('TitanicPrediction.csv', index=False)
TitanicPrediction
```
